```python
import torch
import matplotlib.pyplot as plt
```

```python
# ReLU
x = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.relu(x)

plt.figure(figsize=[5, 2.5])
plt.grid(True)
plt.xlabel('x')
plt.ylabel('relu(x)')
plt.plot(x.detach().numpy(), y.detach().numpy())
plt.show()
```

```python
# grad of relu
# 有用的神经元激活，无用的直接消除，同时由于grad非0即1，避免了梯度消失
y.backward(torch.ones_like(y), retain_graph=True)

plt.figure(figsize=[5, 2.5])
plt.grid(True)
plt.xlabel('x')
plt.ylabel('grad of relu')
plt.plot(x.detach().numpy(), x.grad)
plt.show()
```

```python
# sigmoid
x = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.sigmoid(x)

plt.figure(figsize=[5, 2.5])
plt.grid(True)
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.plot(x.detach().numpy(), y.detach().numpy())
plt.show()
```

```python
# grad of sigmoid
y.backward(torch.ones_like(y), retain_graph=True)

plt.figure(figsize=[5, 2.5])
plt.grid(True)
plt.xlabel('x')
plt.ylabel('grid of sigmoid')
plt.plot(x.detach().numpy(), x.grad)
plt.show()
```
